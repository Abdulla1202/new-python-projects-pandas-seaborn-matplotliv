### 1) Insturstion (for data cleaning)
### Find the null value in datasets if there is any null value in any column then fill it with the mean of that column

```python
car.isnull().sum()
```

```python
price_cols = ['MSRP', 'Invoice']

for col in price_cols:
    car[col] = (
        car[col]
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
    )
    car[col] = pd.to_numeric(car[col], errors='coerce')
```

```python
numeric_cols = [
    'EngineSize', 'Cylinders', 'Horsepower',
    'MPG_City', 'MPG_Highway', 'Weight',
    'Wheelbase', 'Length', 'MSRP', 'Invoice'
]

for col in numeric_cols:
    car[col] = car[col].fillna(car[col].mean())
```

```python
string_cols = ['Make', 'Model', 'Type', 'Origin', 'DriveTrain']

car.loc[:, string_cols] = car[string_cols].fillna('Unknown')

```
